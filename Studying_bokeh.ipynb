```
# ready
import numpy as np
import pandas as pd
import pandas_bokeh
```

```
# load random data 
df = pd.DataFrame({
    'year' : np.arange(2010,2020),
    'category-a' : np.random.uniform(9000, 15000, 10),
    'category-b' : np.random.uniform(9000, 15000, 10),
    'category-c' : np.random.uniform(9000, 15000, 10)
})
df
```

```
# bokeh set up
pandas_bokeh.output_notebook()
```

```
# 막대그래프 bar
df.plot_bokeh(
    kind='bar',
    # data 설정
    x='year',
    y=['category-a', 'category-b', 'category-c'],
    # label
    xlabel='Category',
    ylabel='Annual Sales',
    title='Annual Sales by Category'
)
```

```
pandas_bokeh.output_file('chart0722.html')
```

-> pandas_bokeh.output_notebook()  
-> plot 그림  
-> pandas_bokeh.output_file('chart0722.html')  
하면 html 바로 저장됨

```
# 히스토그램 hist
df.plot_bokeh(
    kind='hist',
    # data 설정
    x='year',
    y=['category-a', 'category-b', 'category-c'],
    # label
    xlabel='Category',
    ylabel='Annual Sales',
    title='Annual Sales by Category'
)
```

```
# 시계열 line
df.plot_bokeh(
    kind='line',
    # data 설정
    x='year',
    y=['category-a', 'category-b', 'category-c'],
    # label
    xlabel='Category',
    ylabel='Annual Sales',
    title='Annual Sales by Category'
)
```

```
# line 그래프, 옵셜설정
save = df.plot_bokeh(
      kind='line',
      # data 설정
      x='year',
      y=['category-a', 'category-b', 'category-c'],
      # label
      xlabel='Category',
      ylabel='Annual Sales',
      title='Annual Sales by Category',
      # 옵션
      zooming=False,
      panning=False
  )
```
